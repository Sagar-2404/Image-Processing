```python
#Program for gray level slicing with and without background

from PIL import Image
from IPython.display import display
import numpy as np
```

```python
def actual_image():
    image = Image.open("Fig0222(b)(cameraman).tif")
    display(image)
def without_background():
    image = Image.open("Fig0222(b)(cameraman).tif")
    img_array = np.array(image,dtype=np.uint8)
    img_shape=(image.width,image.height)

    for i in range(img_shape[1]):
        for j in range(img_shape[0]):
            if(img_array[i][j]>0 and img_array[i][j]<50):
                img_array[i][j]=255
            else:
                img_array[i][j]=0

    resultImage = Image.fromarray(img_array)
    display(resultImage)
def with_background():
    image = Image.open("Fig0222(b)(cameraman).tif")
    img_array = np.array(image,dtype=np.uint8)
    img_shape=(image.width,image.height)

    for i in range(img_shape[1]):
        for j in range(img_shape[0]):
            if(img_array[i][j]>0 and img_array[i][j]<50):
                img_array[i][j]=255

    resultImage = Image.fromarray(img_array)
    display(resultImage)
```
