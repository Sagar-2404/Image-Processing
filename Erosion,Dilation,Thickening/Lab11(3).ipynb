```python
from PIL import Image
import numpy as np
import os

def img_to_arr(img_path):
    with Image.open(img_path) as img:
        img_arr = np.array(img, dtype=np.uint8)
    return img_arr

def arr_to_img(img_new, save_path):
    result_img = Image.fromarray(img_new*255)
    result_img.save(save_path)

def erode(img_arr, se):
    irow, icol = img_arr.shape[0], img_arr.shape[1]
    img_new = np.zeros((irow, icol), dtype=np.uint8)
    mrow, mcol = se.shape[0], se.shape[1]
    for i in range(mrow//2, irow-mrow//2):
        for j in range(mcol//2, icol-mcol//2):
            img_new[i, j] = 1
            for k in range(mrow):
                for l in range(mcol):
                    if se[k][l]==1 and img_arr[i-mrow//2+k, j-mcol//2+l]==0:
                        img_new[i, j] = 0
                        break
                else:
                    continue
                break
    return img_new

def dilate(img_arr, se):
    irow, icol = img_arr.shape[0], img_arr.shape[1]
    img_new = np.zeros((irow, icol), dtype=np.uint8)
    mrow, mcol = se.shape[0], se.shape[1]
    for i in range(mrow//2, irow-mrow//2):
        for j in range(mcol//2, icol-mcol//2):
            img_new[i, j] = 0
            for k in range(mrow):
                for l in range(mcol):
                    if se[k][l]==1 and img_arr[i-mrow//2+k, j-mcol//2+l]==1:
                        img_new[i, j] = 1
                        break
                else:
                    continue
                break
    return img_new

def erosion(img_path, save_path, se):
    arr_to_img(erode(img_to_arr(img_path), se), save_path)

def dilation(img_path, save_path, se):
    arr_to_img(dilate(img_to_arr(img_path), se), save_path)

def opening(img_path, save_path, se):
    arr_to_img(dilate(erode(img_to_arr(img_path), se), se), save_path)

def closing(img_path, save_path, se):
    arr_to_img(erode(dilate(img_to_arr(img_path), se), se), save_path)

se = np.ones((5,5), dtype=np.uint8)
erosion('Fig0905(a)(wirebond-mask).tif','erosion1.tif', se)
dilation('Fig0905(a)(wirebond-mask).tif','dilation1.tif', se)
opening('Fig0905(a)(wirebond-mask).tif','opening.tif', se)
closing('Fig0905(a)(wirebond-mask).tif','closing.tif', se)
```
