```python
from PIL import Image
import numpy as np

def img_to_arr(img_path):
    with Image.open(img_path) as img:
        img_arr = np.array(img, dtype=np.uint8)
    return img_arr

def arr_to_img(img_new, save_path):
    img_new = np.where(img_new==1, 255, img_new)
    result_img = Image.fromarray(img_new)
    result_img.save(save_path)

def HMT(img_arr, se):
    irow, icol = img_arr.shape[0], img_arr.shape[1]
    img_new = np.zeros((irow, icol), dtype=np.uint8)
    mrow, mcol = se.shape[0], se.shape[1]
    for i in range(mrow//2, irow-mrow//2):
        for j in range(mcol//2, icol-mcol//2):
            img_new[i, j] = 1
            for k in range(mrow):
                for l in range(mcol):
                    if se[k][l]==1 or se[k][l]==0:
                        if se[k][l]!=img_arr[i-mrow//2+k, j-mcol//2+l]:
                            img_new[i, j] = 0
                            break
                else:
                    continue
                break
    return img_new

def thinning(img_path, save_path, ses):
    img_new = img_to_arr(img_path)
    for _ in range(10):
        for se in ses:
            img_new = img_new & ~HMT(img_new, se)
    arr_to_img(img_new, save_path)

def thickening(img_path, save_path, ses):
    img_new = img_to_arr(img_path)
    for _ in range(3):
        for se in ses:
            img_new = img_new | HMT(img_new, se)
    arr_to_img(img_new, save_path)

se1 = np.array([[0,0,0],[-1,1,-1],[1,1,1]])
se2 = np.array([[-1,0,0],[1,1,0],[1,1,-1]])
se3 = np.array([[1,-1,0],[1,1,0],[1,-1,0]])
se4 = np.array([[1,1,-1],[1,1,0],[-1,0,0]])
se5 = np.array([[1,1,1],[-1,1,-1],[0,0,0]])
se6 = np.array([[-1,1,1],[0,1,1],[0,0,-1]])
se7 = np.array([[0,-1,1],[0,1,1],[0,-1,1]])
se8 = np.array([[0,0,-1],[0,1,1],[-1,1,1]])
se = [se1,se2,se3,se4,se5,se6,se7,se8]
thinning('Fig0905(a)(wirebond-mask).tif','thinning.tif', se)
se1 = np.array([[1,1,1],[-1,0,-1],[0,0,0]])
se2 = np.array([[-1,1,1],[0,0,1],[0,0,-1]])
se3 = np.array([[0,-1,1],[0,0,1],[0,-1,1]])
se4 = np.array([[0,0,-1],[0,0,1],[-1,1,1]])
se5 = np.array([[0,0,0],[-1,0,-1],[1,1,1]])
se6 = np.array([[-1,0,0],[1,0,0],[1,1,-1]])
se7 = np.array([[1,-1,0],[1,0,0],[1,-1,0]])
se8 = np.array([[1,1,-1],[1,0,0],[-1,0,0]])
se = [se1,se2,se3,se4,se5,se6,se7,se8]
thickening('Fig0905(a)(wirebond-mask).tif','thickening.tif', se)
```
